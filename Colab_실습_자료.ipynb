1. __확정적 조사 (_conclusive study_)__  
+ 문제를 해결하거나 의사결정에 필요한 최종 정보를 얻기 위해 수행  
+ 기술적 조사와 인과관계 조사</br>
2. __기술적 조사 (_descriptive study_)__  
+ 어떠한 사건이나 현상에 기초한 사실적 연구 (중범위 연구)  
+ 현황을 전체적으로 파악하고, 관련 변수간의 상호관계성을 파악  
+ 연구자의 주관이 최대한 절제</br>
3. __인과관계 조사 (_casual study, explanatory study_)__  
+ 현상 내 핵심 요인 간의 인과관계 규명 (진단적 연구)  
+ 미래의 사실에 대한 예측 (예측적 연구)  
+ 연구자의 해석과 주관이 많이 개입 (미시적 연구)
